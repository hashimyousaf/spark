```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("FriendsByAge")
sc = SparkContext(conf = conf)
```

```python


def parseLine(line):
    fields = line.split(',')
    age = int(fields[2])
    numFriends = int(fields[3])
    return (age, numFriends)
```

```python

lines = sc.textFile("file:///home/hashimyousaf/spark-2.4.0-bin-hadoop2.7/bin/jupyter-scripts/dataset/friends/fakefriends.csv")
```

```python
rdd = lines.map(parseLine)
print(rdd.collect())
```

```python

totalsByAge = rdd.mapValues(lambda x: (x, 1))
print(totalsByAge.collect())

```

```python
totalsByAge = totalsByAge.reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
print(totalsByAge.collect())
```

```python
averagesByAge = totalsByAge.mapValues(lambda x: x[0] / x[1])
print(averagesByAge.collect())
```

```python

results = averagesByAge.collect()
for result in results:
    print(result)
```
